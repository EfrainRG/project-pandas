```python
import pandas as pd
import re
```

```python
# Tuve un problema a la hora leer el csv de la data, el cual fue:
# UnicodeDecodeError: 'utf-8' codec can't decode byte 0x92 in position 1: invalid start byte
# asi que tuve que ocupar una codifciaci√≥n diferente
data = pd.read_csv('GSAF5.csv', encoding = "ISO-8859-1")
```

```python
data
```

```python
# primero obtenemos los headers de las columnas actuales
columns = data.columns
data.columns
```

```python
# analizando el csv, eliminamos las columnas con informaci√≥n duplicada
# se observa tambi√©n que hay dos columnas inservible, adem√°s de la columna 
# de pdf ya que ese nombre de pdf puede ser consultado en la URL
data = data.drop(['href','Case Number.2','Unnamed: 22','Unnamed: 23','pdf'], axis = 1)
```

```python
data
```

```python
# Vamos a renombrar para que sea m√°s f√°cil el etendimiento de la columna
data = data.rename(columns={'href formula':'URL of PDF',
                            'Displacement':'Engine Displacement',
                            'Sex ':'Sex',
                           'Fatal (Y/N)':'Fatal (Y/N/Unknown)',
                           'Species ':'Species',
                           'original order':'Original Order'})
```

```python
# Vamos a reordenar las columnas
column_order = ['Original Order','Case Number.1','Case Number','Date','Year','Time',
               'Country','Area','Location','Name','Age','Sex','Activity','Type',
                'Injury','Fatal (Y/N/Unknown)','Species','Investigator or Source','URL of PDF']
```

```python
data = data[column_order]
data
```

```python
# vamos a limpiar la columna de 'Date', quitamos la cadena 'Reported ' para dejar unicamente la fecha
data['Date'] = [x.replace('Reported', '') for x in data['Date']]
```

```python
# Quitaremos los espacios en blanco con los que empiezan en 'Date'
data['Date'] = [x.lstrip() for x in data['Date']]
```

```python
# Quitaremos los espacios en blanco con los que empiezan en 'Date'
data['Date'] = [x.lstrip() for x in data['Date']]
```

```python
# En columna 'Fatal (Y/N)' reemplazaremos valores en blanco o desconocidos
# Intente meter un if para cualqueri valro diferente a 'Y' o 'N' coloque un 'UNKNOW' pero no lo logr√©
data['Fatal (Y/N/Unknown)'] = data['Fatal (Y/N/Unknown)'].str.replace('F','UNKNOW')
```

```python
data['Fatal (Y/N/Unknown)'] = data['Fatal (Y/N/Unknown)'].str.replace(' ','UNKNOW')
```

```python
data['Fatal (Y/N/Unknown)'] = data['Fatal (Y/N/Unknown)'].str.replace('#VALUE!','UNKNOW')
```

```python
data
```

```python
# Trabajaremos con la columna 'Time'
```

```python
data['Time'] = data['Time'].str.replace(' ',' UNKNOW')
```

```python
data = data.drop(['Time_extract'], axis = 1)
```

```python
data['Time'] = data['Time'].str.replace('""','')
```

```python
data['Time'] = data['Time'].str.replace('>','')
```

```python
data['Time'] = data['Time'].str.replace('h',':')
```

```python
# Se extraen los n√∫meros de 'Time' para ponerlos en una nueva columna
data['Time_extract'] = data['Time'].str.extract('(\d+:\d+)')
#data
```

```python
# Tratar de encontrar los nombres, todav√≠a se debe perfecionar
data['Name_extract'] = data['Name'].str.extract('([A-Z][a-z]+.[A-Z][a-z]+)')
data
```
